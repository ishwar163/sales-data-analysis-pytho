# ðŸ§  Sales Data Analysis Project
A beginner-friendly project using Pandas, NumPy, and Seaborn.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

## ðŸ“¥ Load Dataset

```
df = pd.read_csv('sales_data.csv')
df.head()
```

## ðŸ§¹ Data Cleaning & New Column

```
df['TotalSales'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount'])
df.describe()
```

## ðŸ“Š Regional Sales Summary

```
region_sales = df.groupby('Region')['TotalSales'].sum()
region_sales
```

## ðŸ“ˆ Monthly Sales Trend

```
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['TotalSales'].sum()
monthly_sales
```

## ðŸŽ¨ Visualizations

```
plt.figure(figsize=(10,6))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Total Sales by Region')
plt.show()

monthly_sales.plot(kind='line', marker='o', figsize=(10,6), title='Monthly Sales Trend')
plt.show()
```

## ðŸ§  Insights
- Western region has highest sales
- Laptops & Phones top sellers
- Outliers indicate big bulk orders
